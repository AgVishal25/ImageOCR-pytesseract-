```python
import pytesseract
from pytesseract import Output
import cv2
import numpy as np
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

myconfig = r"--psm 6 --oem 3"

def read_text_in_all_orientations(image_path):
    # Load the image
    image = cv2.imread(image_path)
    height, width, _ = image.shape

    # Convert the image to grayscale
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Rotate the image and perform OCR on each rotated image
    for angle in range(0, 360, 30):
        rotated = rotate_image(image, angle)
        data = pytesseract.image_to_data(rotated, config='myconfig', output_type=Output.DICT)
        amount_boxes = len(data['text'])
        for i in range(amount_boxes):
            if float(data['conf'][i]) > 80:
                (x, y, box_width, box_height) = (data['left'][i], data['top'][i], data['width'][i], data['height'][i])
                image = cv2.rectangle(image, (x, y), (x + box_width, y + box_height), (0, 255, 0), 2)
                image = cv2.putText(image, data['text'][i], (x, y + box_height + 20), cv2.FONT_HERSHEY_SIMPLEX, 1.2,
                            (0, 255, 0), 2, cv2.LINE_AA)

    # Resizing the image to fit within the window
    max_width = 1000
    max_height = 800
    if width > max_width or height > max_height:
        scale = min(max_width / width, max_height / height)
        image = cv2.resize(image, None, fx=scale, fy=scale)

    cv2.imshow("Image with Text Detection", image)
    cv2.waitKey(0)

def rotate_image(image, angle):
    # Get the image center coordinates
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)

    # Perform the rotation
    M = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated = cv2.warpAffine(image, M, (w, h))

    return rotated

# Specify the path to the input image
image_path = 'input1.png'

# Read text in all orientations and perform text detection
read_text_in_all_orientations(image_path)
```

```python
import cv2
import pytesseract

# Path to the Tesseract executable
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

def draw_boxes_and_annotate(image, boxes, texts):
    for i, box in enumerate(boxes):
        x, y, w, h = box
        cv2.rectangle(image, (x, y), (w, h), (0, 255, 0), 2)
        cv2.putText(image, texts[i], (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 0, 255), 2)

def detect_text_in_image(image_path, lang='eng'):
    # Load the input image
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Perform text detection
    boxes = pytesseract.image_to_boxes(gray, lang=lang).splitlines()
    texts = pytesseract.image_to_string(gray, lang=lang, config='--psm 6').splitlines()

    # Process the boxes data
    processed_boxes = []
    for box in boxes:
        _, left, top, width, height, _ = box.split()
        processed_boxes.append((int(left), int(top), int(width), int(height)))

    # Draw bounding boxes and annotate the image
    draw_boxes_and_annotate(image, processed_boxes, texts)

    # Show the annotated image
    cv2.imshow('Text Detection', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Example usage
image_path = 'input1.png'
detect_text_in_image(image_path, lang='eng')
```
